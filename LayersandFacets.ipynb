```python
import polars as pl
import plotnine as pn
```

```python
toy_data = pl.DataFrame({"A": (1, 2, 4, 9),
"B": (1, 4, 15, 80),
"C": (1,2,3,4),
"D": ('far', 'far', 'near', 'near')})
toy_data
```

```python
(
    pn.ggplot(toy_data, pn.aes(x="A", y="B"))
    + pn.geom_line(group=1)
    + pn.geom_point(pn.aes(color="D", shape="D"))
)
```

```python
(
    pn.ggplot(toy_data, pn.aes(x="A", y="B"))
    + pn.geom_line(pn.aes(color="D", group="D"), size=1.2)
    + pn.geom_point(pn.aes(color="D", shape="D"), size=4)
)
```

```python
(
    pn.ggplot(toy_data, pn.aes(x="A", y="B"))
    + pn.geom_line(pn.aes(group="D"), color="black", size=1.2)
    + pn.geom_point(pn.aes(color="", shape="D"), size=4)
)
```

```python
(
    pn.ggplot(toy_data, pn.aes(x="A", y="B"))
    + pn.geom_line(pn.aes(group="D", linetype="D"), size=1.2)
    + pn.geom_point(pn.aes(color="D", shape="D"), size=4)
)
```

```python
#2.1.4 
import polars as pl
import plotnine as pn

mexico_hc_spending = pl.read_csv(
    "https://github.com/yardsale8/DSCI_210_R_notebooks/raw/main/data/Mexico_Health_Spending.csv"
)
mexico_hc_spending.head()
```

```python
mexico_hc_spending = mexico_hc_spending.rename({
    "Spending (in billions, 2017 Purchasing Power Parity Dollars PPP $)": "Spending"
})
```

```python
(
    pn.ggplot(mexico_hc_spending, pn.aes(x="Year", y="Spending", fill="Type"))
    + pn.geom_bar(stat="identity")
    + pn.facet_grid("Type ~ .")
    + pn.labs(x="Year", y="Spending")
)
```
